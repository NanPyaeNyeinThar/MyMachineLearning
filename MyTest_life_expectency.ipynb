```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
```

Creation of *data* set by doing selection of columns


```python
# Replace 'your_file.xlsx' with the actual path to your Excel file
df= pd.read_csv("https://raw.githubusercontent.com/NanPyaeNyeinThar/MyMachineLearning/main/life%20expectancy.csv")
```

```python
data=df[df['Year']>=2010]
data=df[(df['Region'] == "East Asia & Pacific")|(df['Region'] == "south Asia")|(df['Region']== "Europe & Central Asia")]
data1= data[['Health Expenditure %','Education Expenditure %', 'Life Expectancy World Bank']]
data1
```

Preping the data

```python
#For Life Expectancy World Bank
# Separate the numeric and non-numeric values
numeric_values = data1['Health Expenditure %'].loc[pd.to_numeric(data1['Health Expenditure %'], errors='coerce').notna()]
non_numeric_values = data1['Health Expenditure %'].loc[pd.to_numeric(data1['Health Expenditure %'], errors='coerce').isna()]

# Calculate the most frequent numeric value
most_frequent_numeric = numeric_values.mode().iloc[0]

# Replace missing numeric values with the most frequent numeric value
data1['Health Expenditure %'] = pd.to_numeric(data1['Health Expenditure %'], errors='coerce').fillna(most_frequent_numeric)

# Replace missing non-numeric values with a default value (e.g., 'Unknown')
data1['Health Expenditure %'].fillna('Unknown', inplace=True)

data1
```

```python
# For Education Expenditure %
# Separate the numeric and non-numeric values
numeric_values = data1['Education Expenditure %'].loc[pd.to_numeric(data1['Education Expenditure %'], errors='coerce').notna()]
non_numeric_values = data1['Education Expenditure %'].loc[pd.to_numeric(data1['Education Expenditure %'], errors='coerce').isna()]

# Calculate the most frequent numeric value
most_frequent_numeric = numeric_values.mode().iloc[0]

# Replace missing numeric values with the most frequent numeric value
data1['Education Expenditure %'] = pd.to_numeric(data1['Education Expenditure %'], errors='coerce').fillna(most_frequent_numeric)

# Replace missing non-numeric values with a default value (e.g., 'Unknown')
data1['Education Expenditure %'].fillna('Unknown', inplace=True)

data1
```

```python
#Life Expectancy World Bank
# Separate the numeric and non-numeric values
numeric_values = data1['Life Expectancy World Bank'].loc[pd.to_numeric(data1['Life Expectancy World Bank'], errors='coerce').notna()]
non_numeric_values = data1['Life Expectancy World Bank'].loc[pd.to_numeric(data1['Life Expectancy World Bank'], errors='coerce').isna()]

# Calculate the most frequent numeric value
most_frequent_numeric = numeric_values.mode().iloc[0]

# Replace missing numeric values with the most frequent numeric value
data1['Life Expectancy World Bank'] = pd.to_numeric(data1['Life Expectancy World Bank'], errors='coerce').fillna(most_frequent_numeric)

# Replace missing non-numeric values with a default value (e.g., 'Unknown')
data1['Life Expectancy World Bank'].fillna('Unknown', inplace=True)

data1
```

```python
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def my_linear_regression(X_train,y_train):

  steps = [("imp_mean", SimpleImputer()), # cleaning the data - replace the missing data with average
          ("scale", StandardScaler()), #standardizing the data
         ('transform', PolynomialFeatures(degree=3)),  #poly to linear transform
            ("Linear", LinearRegression()) ] # linear model - ML

  pipeline = Pipeline(steps)
  model = pipeline.fit(X_train, y_train)
  linear_regressor = pipeline.named_steps['Linear']

  return model, linear_regressor.intercept_, linear_regressor.coef_
```
